```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="day", style="day", palette="deep", size='size')
plt.legend(loc='center', bbox_to_anchor=(1.16, 0.5), title="Day of the Week")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.relplot(x="total_bill", y="tip", data=tips, hue="day", style="day", palette="deep", size='size', col='time')
plt.show()
```
