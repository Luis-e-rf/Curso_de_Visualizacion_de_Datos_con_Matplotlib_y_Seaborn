```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
data
```

## Histograma

```python
plt.hist(data,bins=20,histtype='bar')
plt.show()
```

```python
plt.hist(data,bins=10,histtype='step')
plt.show()
```

## Boxplot

```python
data = np.append(data,200)
plt.boxplot(data,vert=False,patch_artist=True,notch=True)
plt.show()
```

```python
plt.boxplot(data,vert=False,patch_artist=True,notch=True,showfliers=False)
plt.show()
```

## Scatter o dispersion

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) ** 2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y,s=area,c=colors, marker='v',alpha=0.5)
plt.show()
```
