```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')

tips.head()
```

```python
sns.histplot(data=tips,x='tip')
plt.show()
```

```python
sns.histplot(data=tips,x='tip',cumulative=True)
plt.show()
```

```python
sns.histplot(data=tips,x='tip',cumulative=True,hue='sex')
plt.show()
```

```python
sns.histplot(data=tips,x='tip',bins=30,cumulative=False,hue='sex',stat='percent')
plt.show()
```

```python
sns.histplot(data=tips,x='tip',bins=30,cumulative=False,hue='sex',stat='count',multiple='dodge')
plt.show()
```

```python
sns.kdeplot(data=tips,x='tip',hue='sex',fill=True,bw_adjust=1)
plt.show()
```

```python
sns.ecdfplot(data=tips,x='tip',hue='sex',stat='count')
plt.show()
```

```python
sns.displot(data=tips,x='tip',hue='sex',kind='kde')
plt.show()
```
