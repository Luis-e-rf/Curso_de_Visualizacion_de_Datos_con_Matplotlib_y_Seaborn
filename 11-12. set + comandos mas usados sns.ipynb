```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
```

```python
sns.set_theme(style='darkgrid',palette='deep', font='Verdana', font_scale=1.1)
sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
```

# Comandos mas usados

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.displot(data=tips,x='total_bill')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',y='tip')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',y='tip',hue='sex')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',hue='sex',kind='kde',legend=False,palette='dark')
plt.show()
```
